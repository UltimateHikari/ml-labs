## Метод k-средних

Метод k-средних (англ. k-means) — наиболее популярный метод кластеризации. Был изобретён в 1950-х годах математиком Гуго Штейнгаузом и почти одновременно Стюартом Ллойдом. Особую популярность приобрёл после работы Маккуина. Алгоритм стремится минимизировать суммарное квадратичное отклонение точек кластеров от центров этих кластеров.

## Подготовка

```python
import os
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
import warnings

random_state = 42
np.random.seed(random_state)

warnings.simplefilter('ignore')
np.set_printoptions(precision=4, suppress=True)
sns.set_theme()
```

```python
def make_pretty(styler):
    styler.set_caption("Weather Conditions")
    styler.background_gradient(axis=None, vmin=1, vmax=5, cmap="YlGnBu")
    return styler
```

```python
df = pd.read_csv("./Live_20210128.csv")
df.head(5).style.pipe(make_pretty)
```

```python
df.shape
```

## Чистка и перекодирование

```python
df.drop(
    ['Column1', 'Column2', 'Column3', 'Column4', 'status_id', 'status_published'],
    axis=1, inplace=True)
```

```python
df['status_type'] = LabelEncoder().fit_transform(df['status_type'])
```

```python
df.describe().T.style.pipe(make_pretty)
```

```python
df.head(5).style.pipe(make_pretty)
```

```python
X = df.values

X = MinMaxScaler().fit_transform(X)
```

```python
sns.clustermap(df.corr(),annot=True)
```

## Модель
Подберем параметры по силуэту

```python
def cv_score(model, X):
    model.fit(X)
    labels = model.labels_
    num_labels = len(set(labels))
    if num_labels == 1:
        raise ValueError('number of labels must be greater than 2')
    else:
        return silhouette_score(X, labels)
```

```python
param = {"n_clusters": [2, 3, 4, 5, 6, 7, 8, 9, 10, 15, 20, 30]}
```

```python
tun = GridSearchCV(estimator=KMeans(),
                       param_grid=param,
                       scoring=cv_score,
                       cv=2,
                       n_jobs=-1)

tun.fit(X)
```

```python
pd.DataFrame(tun.cv_results_).style.pipe(make_pretty)
```

```python
model = tun.best_estimator_
model.fit(X)
```

## Результат

```python
plt.scatter(X[:, 1], X[:, 2], c=model.predict(X), s=50, cmap='viridis')
plt.scatter(model.cluster_centers_[1], 
            model.cluster_centers_[2], c='black', s=700, alpha=0.5)
```
